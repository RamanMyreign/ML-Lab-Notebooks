```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
data=pd.read_csv('Missing data 28-08-2020.csv')
data.head()
```

```python
from sklearn.impute import SimpleImputer
replace_nan =  SimpleImputer(missing_values=np.nan,strategy='most_frequent')
columns = data.columns
data = pd.DataFrame(replace_nan.fit_transform(data))
data.columns = columns
data
```
